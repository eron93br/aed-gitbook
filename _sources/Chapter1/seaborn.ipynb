# Seaborn

Seaborn é uma biblioteca de visualização de dados em Python baseada no Matplotlib, projetada para criar visualizações estatísticas atraentes e informativas. **Ela fornece uma interface de alto nível que simplifica a criação de gráficos complexos**, permitindo aos usuários focar na exploração e na compreensão dos dados.

Uma das principais vantagens do Seaborn é a sua capacidade de criar gráficos estatísticos complexos com poucas linhas de código. Ele oferece suporte a uma variedade de gráficos estatísticos, incluindo gráficos de dispersão com ajuste de regressão, gráficos de distribuição de variáveis univariadas e bivariadas, gráficos de categorias, gráficos de séries temporais e muito mais.

Além dos gráficos estatísticos, o Seaborn também permite personalizar diversos aspectos visuais dos gráficos, como cores, estilos, títulos, rótulos de eixos e legendas. Ele também possui temas pré-definidos que melhoram a aparência dos gráficos, tornando-os mais atraentes e profissionais.

Outra característica interessante do Seaborn é a sua integração com as estruturas de dados do Pandas, outra biblioteca popular em Python para análise de dados. O Seaborn pode trabalhar diretamente com DataFrames do Pandas, facilitando a manipulação e visualização de dados tabulares.

```python
import seaborn as sns
import pandas as pd
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv')
```

```python
# usando seaborn para ver quais colunas possuem valores nulos
sns.heatmap(titanic.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
titanic.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'],axis=1,inplace=True)
```

### Countplot

```python
sns.countplot(x='Survived', data=titanic, palette='RdBu_r');
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Sex', hue='Survived',data=titanic,palette='RdBu_r')
```

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic, palette='Blues')
```

### Histograma

```python
sns.histplot(data=titanic, x="Age")
```

```python
sns.histplot(data=titanic, x="Age", kde=True)
```

```python
sns.histplot(data=titanic, x="Age", hue="Survived", kde=True)
```
