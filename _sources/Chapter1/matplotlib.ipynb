O Matplotlib é uma biblioteca de visualização de dados em Python que fornece uma variedade de ferramentas para criar gráficos, plots e visualizações de alta qualidade. É uma das bibliotecas mais populares para visualização de dados em Python devido à sua facilidade de uso e à vasta gama de recursos que oferece.

A biblioteca Matplotlib foi criada por John D. Hunter em 2003 com o objetivo de produzir gráficos semelhantes aos do MATLAB, uma popular linguagem de programação usada para computação numérica. Desde então, o Matplotlib evoluiu para se tornar uma ferramenta poderosa e flexível para visualização de dados em Python.

Uma das características mais úteis do Matplotlib é a sua capacidade de personalização. É possível ajustar praticamente todos os aspectos de um gráfico, como títulos, rótulos de eixos, cores, estilos de linha, marcadores e legendas. Além disso, o Matplotlib suporta a criação de gráficos com múltiplas subplots, permitindo combinar várias visualizações em uma única figura.

## Como criar um gráfico

Basicamente, temos duas formas diferentes de criar gráficos. A função ```plot()``` usa parâmetros para especificar pontos no diagrama. O primeiro parâmetro é uma matriz contendo os pontos no eixo x. O segundo é uma matriz contendo os pontos no eixo y.


````python
x = [1, 2, 3, 4]
y = [1212, 333, 9931, -1212]
````

````python
# FORMA 1
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
````

````python
# FORMA 2
plt.plot(x, y)
````

Porém note que podemos alterar a estrutura do gráfico

````python
plt.plot(x, y, 'o')
plt.show()
````

Também podemos limitar as visualizações nos eixos x e y!

````python
plt.plot(x,y)
plt.axis([0, 5000, 0, 35540])
````

Um resumo destes marcadores pode ser listado abaixo:

1. **Marcadores básicos**:

- ```.``` (Período): Marcador de pontos.
- ```,``` (vírgula): marcador de pixel.
- ```o``` (Círculo): Marcador de círculo.
- ```v``` (triângulo para baixo): marcador de triângulo apontando para baixo.
- ```^``` (Triângulo para cima): marcador de triângulo apontando para cima.
- ```<``` (Triângulo Esquerdo): Marcador de triângulo apontando para a esquerda.
- ```>``` (Triângulo à direita): Marcador de triângulo apontando para a direita.
- ```s``` (Quadrado): Marcador quadrado.
- ```d``` (Diamante): marcador de diamante.
- ```p``` (Pentágono): marcador do Pentágono.
- ```*``` (Estrela): Marcador de estrela.
- ```h``` (Hexagon1): Marcador de hexágono (1).
- ```H``` (Hexagon2): Marcador de hexágono (2).
- ```+``` (Mais): Marcador de mais.
- ```x``` (Cruz): marcador de cruz.

2. **Marcadores Avançados**:

- ```1``` (Tri Down): marcador de triângulo apontando para baixo.
- ```2``` (Tri Up): marcador de triângulo apontando para cima.
- ```3``` (Tri Esquerda): Marcador de triângulo apontando para a esquerda.
- ```4``` (Tri Right): Marcador de triângulo apontando para a direita.
- ```8``` (Octógono): marcador de octógono.
- ```s``` (Quadrado): Marcador quadrado.
- ```p``` (Pentágono): marcador do Pentágono.
- ```P``` (Mais (preenchido)): Preenchido com marcador.
- ```X``` (X (preenchido)): Marcador X preenchido.
- ```|``` (Vline): Marcador de linha vertical.
- ```_``` (Hline): Marcador de linha horizontal.

3. **Marcadores personalizados**:

É possível criar marcadores personalizados no Matplotlib usando objetos de caminho (Path). Isso possibilita aos usuários definirem suas próprias formas de marcador, combinando diferentes comandos de caminho e proporcionando flexibilidade na criação de marcadores únicos.

## Integração do Pandas com Matplotlib

Podemos definir o tipo de gráfico que desejamos trabalhar de duas formas. A primeira delas é passando o argumento kind dentro da função plot(), assim:

```df.plot(x, y, kind='bar')```

Outra forma de ser feito é utilizando a função específica para o gráfico desejado. Nesse exemplo em específico, seria:

 ```plt.bar(x,y)```

 ### Algumas dicas

- ```bar``` ou ```barh``` para bar plots

- ```hist``` para histograma

- ```box``` para boxplot

- ```kde``` ou ```density``` para density plots

### Histogramas

````python
# lista para realizar o grafico
horas = [17, 20, 22, 25, 26, 27, 30, 31, 32, 38, 40, 40, 45, 55]

# incializacao
fig, ax = plt.subplots(figsize = (6, 4))

# plot
ax.hist(horas, bins=5, edgecolor="black");
````

````python
df['Quantity'].hist()
````

````python
df['Payment'].value_counts().plot.hist()
````

````python
# Inicializacao do grafico
fig, ax = plt.subplots(figsize = (10, 4))

# realizar o grafico 
ax.hist(df['Rating'][:8], bins=5, edgecolor="black");
ax.set_title('Distribuicao das avaliacoes de compra')
ax.set_ylabel('Frequencia')
ax.set_xlabel('Nota da Avaliacao')
````

### Grafico de Pizza

Divisão por Tipo de Pagamento

````python
# figure(figsize=(8, 6))
plot1 = plt.figure(figsize=(12,7))
gender = df.Payment.value_counts()
labels = df.Payment.unique()
plt.pie(gender,labels=labels,autopct='%.0f%%')
plt.title("Formas de pagamento utilizadas")
plt.show()
````

### Graficos de Barra

````python
# lista com valores
valores = [3, 12, 5, 18, 45]
barras = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(barras))
 
# Cria barras horizontais
plt.barh(y_pos, valores)
 
# Crria o nome da legenda no eixo x
plt.yticks(y_pos, barras)
 
# visualiza o grafico
plt.show()
````

````python
figbar = plt.figure(figsize=(10,6))
df['Payment'].value_counts().plot.bar()
plt.xticks(rotation=0)
plt.title("Formas de pagamento utilizadas")
plt.ylabel("Numero de Pagamentos")
````
