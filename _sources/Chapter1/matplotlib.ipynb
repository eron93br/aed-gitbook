O Matplotlib  uma biblioteca de visualizao de dados em Python que fornece uma variedade de ferramentas para criar grficos, plots e visualizaes de alta qualidade.  uma das bibliotecas mais populares para visualizao de dados em Python devido  sua facilidade de uso e  vasta gama de recursos que oferece.

A biblioteca Matplotlib foi criada por John D. Hunter em 2003 com o objetivo de produzir grficos semelhantes aos do MATLAB, uma popular linguagem de programao usada para computao numrica. Desde ento, o Matplotlib evoluiu para se tornar uma ferramenta poderosa e flexvel para visualizao de dados em Python.

Uma das caractersticas mais teis do Matplotlib  a sua capacidade de personalizao.  possvel ajustar praticamente todos os aspectos de um grfico, como ttulos, rtulos de eixos, cores, estilos de linha, marcadores e legendas. Alm disso, o Matplotlib suporta a criao de grficos com mltiplas subplots, permitindo combinar vrias visualizaes em uma nica figura.

## Como criar um grfico

Basicamente, temos duas formas diferentes de criar grficos. A funo ```plot()``` usa parmetros para especificar pontos no diagrama. O primeiro parmetro  uma matriz contendo os pontos no eixo x. O segundo  uma matriz contendo os pontos no eixo y.


````python
x = [1, 2, 3, 4]
y = [1212, 333, 9931, -1212]
````

````python
# FORMA 1
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
````

````python
# FORMA 2
plt.plot(x, y)
````

Porm note que podemos alterar a estrutura do grfico

````python
plt.plot(x, y, 'o')
plt.show()
````

Tambm podemos limitar as visualizaes nos eixos x e y!

````python
plt.plot(x,y)
plt.axis([0, 5000, 0, 35540])
````

Um resumo destes marcadores pode ser listado abaixo:

1. **Marcadores bsicos**:

- ```.``` (Perodo): Marcador de pontos.
- ```,``` (vrgula): marcador de pixel.
- ```o``` (Crculo): Marcador de crculo.
- ```v``` (tringulo para baixo): marcador de tringulo apontando para baixo.
- ```^``` (Tringulo para cima): marcador de tringulo apontando para cima.
- ```<``` (Tringulo Esquerdo): Marcador de tringulo apontando para a esquerda.
- ```>``` (Tringulo  direita): Marcador de tringulo apontando para a direita.
- ```s``` (Quadrado): Marcador quadrado.
- ```d``` (Diamante): marcador de diamante.
- ```p``` (Pentgono): marcador do Pentgono.
- ```*``` (Estrela): Marcador de estrela.
- ```h``` (Hexagon1): Marcador de hexgono (1).
- ```H``` (Hexagon2): Marcador de hexgono (2).
- ```+``` (Mais): Marcador de mais.
- ```x``` (Cruz): marcador de cruz.

2. **Marcadores Avanados**:

- ```1``` (Tri Down): marcador de tringulo apontando para baixo.
- ```2``` (Tri Up): marcador de tringulo apontando para cima.
- ```3``` (Tri Esquerda): Marcador de tringulo apontando para a esquerda.
- ```4``` (Tri Right): Marcador de tringulo apontando para a direita.
- ```8``` (Octgono): marcador de octgono.
- ```s``` (Quadrado): Marcador quadrado.
- ```p``` (Pentgono): marcador do Pentgono.
- ```P``` (Mais (preenchido)): Preenchido com marcador.
- ```X``` (X (preenchido)): Marcador X preenchido.
- ```|``` (Vline): Marcador de linha vertical.
- ```_``` (Hline): Marcador de linha horizontal.

3. **Marcadores personalizados**:

 possvel criar marcadores personalizados no Matplotlib usando objetos de caminho (Path). Isso possibilita aos usurios definirem suas prprias formas de marcador, combinando diferentes comandos de caminho e proporcionando flexibilidade na criao de marcadores nicos.

## Integrao do Pandas com Matplotlib

Podemos definir o tipo de grfico que desejamos trabalhar de duas formas. A primeira delas  passando o argumento kind dentro da funo plot(), assim:

```df.plot(x, y, kind='bar')```

Outra forma de ser feito  utilizando a funo especfica para o grfico desejado. Nesse exemplo em especfico, seria:

 ```plt.bar(x,y)```

 ### Algumas dicas

- ```bar``` ou ```barh``` para bar plots

- ```hist``` para histograma

- ```box``` para boxplot

- ```kde``` ou ```density``` para density plots

### Histogramas

````python
# lista para realizar o grafico
horas = [17, 20, 22, 25, 26, 27, 30, 31, 32, 38, 40, 40, 45, 55]

# incializacao
fig, ax = plt.subplots(figsize = (6, 4))

# plot
ax.hist(horas, bins=5, edgecolor="black");
````

````python
df['Quantity'].hist()
````

````python
df['Payment'].value_counts().plot.hist()
````

````python
# Inicializacao do grafico
fig, ax = plt.subplots(figsize = (10, 4))

# realizar o grafico 
ax.hist(df['Rating'][:8], bins=5, edgecolor="black");
ax.set_title('Distribuicao das avaliacoes de compra')
ax.set_ylabel('Frequencia')
ax.set_xlabel('Nota da Avaliacao')
````

### Grafico de Pizza

Diviso por Tipo de Pagamento

````python
# figure(figsize=(8, 6))
plot1 = plt.figure(figsize=(12,7))
gender = df.Payment.value_counts()
labels = df.Payment.unique()
plt.pie(gender,labels=labels,autopct='%.0f%%')
plt.title("Formas de pagamento utilizadas")
plt.show()
````

### Graficos de Barra

````python
# lista com valores
valores = [3, 12, 5, 18, 45]
barras = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(barras))
 
# Cria barras horizontais
plt.barh(y_pos, valores)
 
# Crria o nome da legenda no eixo x
plt.yticks(y_pos, barras)
 
# visualiza o grafico
plt.show()
````

````python
figbar = plt.figure(figsize=(10,6))
df['Payment'].value_counts().plot.bar()
plt.xticks(rotation=0)
plt.title("Formas de pagamento utilizadas")
plt.ylabel("Numero de Pagamentos")
````
